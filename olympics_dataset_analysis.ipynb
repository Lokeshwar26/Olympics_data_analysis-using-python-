```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athl=pd.read_csv('/content/drive/MyDrive/athletes/athlete_events.csv')
region=pd.read_csv('/content/drive/MyDrive/athletes/noc_regions.csv')
```

```python
athl.head(10)
```

```python
region.head(10)
```

```python
athl_df=athl.merge(region,how='left',on='NOC')
athl_df.head(10)
```

```python
athl_df.shape
```

```python
athl_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
```

```python
#top countries
top_10_countries=athl_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#plot top_10
plt.figure(figsize=(12,6))
plt.title('Overall participation by country')
sns.barplot(x=top_10_countries.index,y=top_10_countries,palette='bright')
```

```python
plt.figure(figsize=(12,6))
plt.title('Age distribution of athletes')
plt.xlabel('Age')
plt.ylabel('No of participants')
plt.hist(athl_df.Age,bins=np.arange(10,80,3),color='blue',edgecolor='black');
```

```python
winter_sports=athl_df[athl_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
summer_sports=athl_df[athl_df.Season=='Summer'].Sport.unique()
summer_sports
```

```python
gender_counts=athl_df.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=90,shadow=False);
```

```python
athl_df.Medal.value_counts()
```

```python
female_participants=athl_df[(athl_df.Sex=='F') & (athl_df.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants
```

```python
womenolympics=athl_df[(athl_df.Sex=='F')&(athl_df.Season=='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenolympics,palette='Spectral')
plt.title('Women participants')
```

```python
part=womenolympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('female participants over the time')
```

```python
gold_medals=athl_df[athl_df.Medal=='Gold']
gold_medals
```
